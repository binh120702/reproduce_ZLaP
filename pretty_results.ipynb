```python
import json
import pandas as pd
```

```python
with open("reproduced_results.txt", 'r') as f:
    data = json.load(f)
```

```python
df_results = pd.DataFrame(data)
df_results
```

```python
# Preserve dataset column order
dataset_order = df_results["dataset"].drop_duplicates().tolist()

# List of backbones
backbones = ["RN50_openai", "ViT-B-16_openai"]

for backbone in backbones:
    # Filter by backbone
    df_filtered = df_results[df_results["backbone"] == backbone]

    # Pivot table: setup as rows, dataset as columns
    df_pivot = df_filtered.pivot_table(
        index="setup",
        columns="dataset",
        values="accuracy"
    )

    # Reorder columns to match original dataset order
    df_pivot = df_pivot.reindex(columns=dataset_order).round(1)

    # Save to CSV
    filename = f"reproduced_{backbone}.csv".replace("-", "_").lower()
    df_pivot.to_csv(filename)
    print(f"Saved {filename}")
```
